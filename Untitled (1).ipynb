FIRST we perform WOE and IV and check which features are good predictors for our target variable.

```python
#WOE AND IV FOR CATEGORICAL COLUMNS
def calc_woe_iv(df, feature, target='loan_status'):
    

    
    # 1. Frequency table
    grouped = df.groupby(feature)[target].agg(['count', 'sum'])
    grouped.columns = ['Total', 'Bad']
    grouped['Good'] = grouped['Total'] - grouped['Bad']
    
    # 2. Distribution
    grouped['Dist_Good'] = grouped['Good'] / grouped['Good'].sum()
    grouped['Dist_Bad'] = grouped['Bad'] / grouped['Bad'].sum()
    
    # 3. WOE
    grouped['WOE'] = np.log((grouped['Dist_Good'] + 1e-6) / (grouped['Dist_Bad'] + 1e-6))
    
    # 4. IV
    grouped['IV'] = (grouped['Dist_Good'] - grouped['Dist_Bad']) * grouped['WOE']
    
    iv_total = grouped['IV'].sum()

    # Check acceptance
    if 0.02 <= iv_total <= 0.5:
        status = 'Accepted'
    else:
        status = 'Rejected'
    
    print(grouped.reset_index())
    print(f"IV Total: {iv_total}")
    print(status)

    
```

```python
for col in cat_cols:
    print(f"WOE AND IV TABLE ALONG WITH DECISION FOR {col} is :\n")
    calc_woe_iv(data,col)
```

```python
#WOE AND IV FOR NUMERICAL COLUMNS
def woe_and_iv(column, nbins):
    # Bin the column with qcut first (equal-frequency)
    bin_col = f'binned_{column}'
    try:
        data[bin_col] = pd.qcut(data[column], q=nbins, duplicates='drop')
    except Exception:
        # fallback: equal-width binning if too few unique values
        data[bin_col] = pd.cut(data[column], bins=nbins, duplicates='drop')
    
    # Pivot table to count number of events/non-events in each bin
    df1 = pd.pivot_table(
        data=data, 
        index=bin_col,
        columns='loan_status',
        values='loan_amnt',   # any column to count; here 'loan_amnt'
        aggfunc='count',
        observed=False
    ).fillna(0)
    
    # Identify classes: 1 = default (bad), 0 = fully paid (good)
    default_class = 1
    fully_paid_class = 0

    # Total counts
    default_sum = df1.get(default_class, pd.Series()).sum()
    fully_paid_sum = df1.get(fully_paid_class, pd.Series()).sum()
    
    # Calculate percentages
    epsilon = 1e-10
    df1['% of Events'] = (df1.get(default_class, pd.Series(0)) / (default_sum + epsilon)) * 100
    df1['% of Non Events'] = (df1.get(fully_paid_class, pd.Series(0)) / (fully_paid_sum + epsilon)) * 100
    
    # Calculate WOE
    df1['WOE'] = np.log((df1['% of Non Events'] + epsilon) / (df1['% of Events'] + epsilon))
    
    # Information Value
    df1['IV'] = df1['WOE'] * (df1['% of Non Events'] - df1['% of Events'])
    
    # Total IV (divide by 100 because percentages were used)
    iv_total = df1['IV'].sum() / 100
    
    # Check acceptance
    status = 'Accepted' if 0.02 <= iv_total <= 0.5 else 'Rejected'
    
    # Clean up: drop bin column
    data.drop(columns=[bin_col], inplace=True)
    
    return df1,iv_total,status
```

```python
columns = ['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'annual_inc',
           'dti', 'delinq_2yrs', 'fico_range_low', 'fico_range_high',
           'inq_last_6mths', 'open_acc', 'pub_rec', 'revol_bal', 'revol_util',
           'total_acc', 'total_pymnt', 'total_rec_prncp', 'total_rec_int',
           'recoveries', 'collection_recovery_fee', 'mort_acc',
           'pub_rec_bankruptcies', 'last_pymnt_amnt']

for column in columns:
    table, iv_sum, status = woe_and_iv(column, nbins=10)
    print(f"=== {column} ===")
    print(table)
    print(f"IV sum: {iv_sum:.4f}")
    print(f"Feature status: {status}\n")
```

```python
#function for outlier handling
def handling_outliers(X): #Creating Function for Handling Outliers ( We use Capping Method)
  X_copy = pd.DataFrame(X,columns=trf1.get_feature_names_out())
  for column in accepted_num_cols:
    if not pd.api.types.is_numeric_dtype(X_copy[column]):
          continue
    q1=X_copy[column].quantile(0.25)
    q3=X_copy[column].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    X_copy[column]=np.where(X_copy[column]>upper_bound,upper_bound,X_copy[column])
    X_copy[column]=np.where(X_copy[column]<lower_bound,lower_bound,X_copy[column])
  return X_copy
    
```

```python
trf2=FunctionTransformer(handling_outliers) #Handles Outliers
```

```python
trf3=ColumnTransformer(transformers=[  #Encoding and Scaling
    ('MinMaxSCaling' , MinMaxScaler() ,accepted_num_cols),
    ('OneHotEncoding', OneHotEncoder(sparse_output=False,drop='first') ,accepted_cat_cols)
] 
            ,remainder='passthrough')
```

```python
pipe = Pipeline([ #Creating Pipeline
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
])
```
